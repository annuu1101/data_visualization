```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 5, 12, 8, 6]
```

```python

plt.scatter(x, y)
plt.title('Simple Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

```python
data = [10, 20, 30, 40, 50]
labels = ['A', 'B', 'C', 'D', 'E']

plt.bar(labels, data)
plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
plt.barh(labels, data)
plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

data = [10, 20, 30, 40, 50]
labels = ['A', 'B', 'C', 'D', 'E']
```

```python
fig, axs = plt.subplots(2, 2, figsize=(6, 6), sharex=True)

# Plot data on each subplot
axs[0,0].plot(x, y1, color='blue')
axs[0,0].set_title('Sine Wave')
axs[0,1].plot(x, y2, color='red')
axs[0,1].set_title('Cosine Wave')
axs[1,0].bar(data, labels, color='green')
axs[1,0].set_title('bar 1')
axs[1,1].pie(data, labels = data)
axs[1,1].set_title('Pie')


```

```python

# Generate some data
x = np.linspace(0, 10, 100)
y = 10 ** x

# Create a plot with a logarithmic y-axis
plt.plot(x, y)
plt.yscale('log')

# Set axis labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Logarithmic Scale')

plt.show()
```
